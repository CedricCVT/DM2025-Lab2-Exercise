```python
#imports

import os
import re
import json
import zipfile
import random

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, classification_report, confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import ComplementNB

# just to make results roughly reproducible
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
random.seed(RANDOM_STATE)
```

```python
# V2 : importing the models for my V2

from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
#kaggle competitions download -c dm-lab-2-private-competition
```

```python
# Load CSV + JSON from the competition zip

DATA_ZIP = "dm-lab-2-private-competition.zip"

if not os.path.exists(DATA_ZIP):
    raise FileNotFoundError(
        f"Can't find {DATA_ZIP}. "
        "pls put the Kaggle competition zip in the same folder as this notebook."
    )

with zipfile.ZipFile(DATA_ZIP, "r") as z:
    with z.open("data_identification.csv") as f:
        data_id = pd.read_csv(f)
    with z.open("emotion.csv") as f:
        emotions = pd.read_csv(f)
    with z.open("final_posts.json") as f:
        posts_raw = json.load(f)
    with z.open("samplesubmission.csv") as f:
        sample_sub = pd.read_csv(f)

print("data_identification shape:", data_id.shape)
print("emotion shape:", emotions.shape)
print("len(posts_raw):", len(posts_raw))
print("sample_sub head:")
display(sample_sub.head())
```

```python
# Creating the df :
#Convert final_posts.json into a DataFrame and merge by id

def posts_to_df(posts_list):
    rows = []
    for p in posts_list:
        #  structure is :  {'root': {'_source': {'post' : { ...}}}}
        post_info = p["root"]["_source"]["post"]
        pid = post_info["post_id"]
        text = post_info["text"]
        hashtags = post_info.get("hashtags", [])
        #join hashtags as a simple space-separated string
        rows.append((pid, text, " ".join(hashtags)))
    return pd.DataFrame(rows, columns=["id", "text", "hashtags"])


posts_df = posts_to_df(posts_raw)

print("posts_df shape:", posts_df.shape)
display(posts_df.head())

# merge: data_identification+posts+emotions (labels only exist for train ids)
df = (
    data_id
    .merge(posts_df, on="id", how="left")
    .merge(emotions, on="id", how="left")
)

print(df["split"].value_counts())
display(df.head())
```

```python
# we can split into train and test using the 'split' column

train_df = df[df["split"] == "train"].copy()
test_df = df[df["split"] == "test"].copy()

print("Train size:", train_df.shape)
print("Test size :", test_df.shape)

print("\nTrain label distribution:")
display(train_df["emotion"].value_counts())
```

```python
# lets vizualize some examples for each emotions

for label in sorted(train_df["emotion"].unique()):
    print("=" * 80)
    print("Emotion:", label)
    sample_rows = train_df[train_df["emotion"] == label].sample(2, random_state=RANDOM_STATE)
    for _, row in sample_rows.iterrows():
        print(f"[{row['id']}] {row['text']}")
    print()
```

```python
# we are now going to do some text cleaning/preprocessing like
# removing urls, mentions, and cutting off weird chars (hashtags chars, but we keep words)
# however we are going to keep punctuation because I think it carries a lot
# of emotions ( !!! ??? ...)

URL_PATTERN = re.compile(r"http\S+|www\.\S+")
MENTION_PATTERN = re.compile(r"@\w+")
HASHTAG_PATTERN = re.compile(r"#(\w+)")

def clean_text(text):
    """
    cleaner for tweets/posts:
    - put everything in lowercase
    - remove urls and mentions
    - remove '#' characters but keep hashtag words (#important !!)
    - keep basic punctuation to keep emotions !!
    """
    text= str(text)
    text= text.lower()
    text=URL_PATTERN.sub(" url ", text)
    text = MENTION_PATTERN.sub(" user ", text)
    text= HASHTAG_PATTERN.sub(r"\1", text)
    # keep only letters, digits, and a few punctuation marks
    text = re.sub(r"[^a-z0-9!?,'\".\s]", " ", text)
    text =re.sub(r"\s+", " ", text).strip()
    return text

# combine main text + hashtags, then clean
train_df["text_raw"] =train_df["text"].fillna("") + " " + train_df["hashtags"].fillna("")
test_df["text_raw"] = test_df["text"].fillna("") + " " + test_df["hashtags"].fillna("")

train_df["text_clean"] = train_df["text_raw"].apply(clean_text)
test_df["text_clean"]= test_df["text_raw"].apply(clean_text)

display(train_df[["text_raw", "text_clean"]].head())
```

```python
# some more vizualizaitions to understand data distribution :

# firstly we have to calculate the numeric features we want to see :

for df_part in (train_df, test_df):
    df_part["char_len"] =df_part["text_clean"].str.len()

train_df[["text_clean", "char_len"]].head()

```

```python
plt.figure(figsize=(6,4))
train_df["emotion"].value_counts().sort_values(ascending=False).plot(kind="bar")
plt.title("Label distribution")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
train_df["char_len"].hist(bins=40)
plt.title("Post length (characters)")
plt.xlabel("chars")
plt.ylabel("count")
plt.show()
```

```python
# Create a new additionnal validation split from the training data, as we learnt in the DM course !!

X_train_text, X_valid_text, y_train, y_valid = train_test_split(
    train_df["text_clean"],
    train_df["emotion"],
    test_size=0.15,
    random_state=RANDOM_STATE,
    stratify=train_df["emotion"],
)

print("Train texts:", X_train_text.shape)
print("Valid texts:", X_valid_text.shape)
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
```

```python
'''
My idea is to build a combined feature space, with word unigrams+bigrams
and character n-grams (3 to 5 cause i dont think we need more) to capture patterns like elongated words, laugh
(like WOOOOW, LOOOOOOL, LMAOOOOO, NO WAYYY, WTFFFF, NOOOO... you get it!)
'''
# TF-IDF feature extractors (word + char)

word_vectorizer = TfidfVectorizer(
    analyzer="word",
    ngram_range=(1, 2),      # unigrams+bigrams
    min_df=3,
    max_df=0.90,
    sublinear_tf=True,
)

char_vectorizer = TfidfVectorizer(
    analyzer="char_wb",
    ngram_range=(3, 5),      # character 3-5-grams
    min_df=5,
    sublinear_tf=True,
)

# combine both in 1 big feature space
combined_vectorizer = FeatureUnion([
    ("word_tfidf", word_vectorizer),
    ("char_tfidf", char_vectorizer),
])
```

```python
# V2 : New TF-IDF setups for tuning
# We try longer n grams, hoping the results will be better
# and the computing not too long

word_vectorizer_opt = TfidfVectorizer(
    analyzer="word",
    ngram_range=(1, 3),      # unigrams, bigrams, trigrams
    min_df=2,                # allow rarer words
    max_df=0.98,             # drop super common ones stop words
    sublinear_tf=True,
)

char_vectorizer_opt = TfidfVectorizer(
    analyzer="char_wb",
    ngram_range=(3, 6),      # a bit longer char-ngrams
    min_df=3,
    sublinear_tf=True,
)

combined_vectorizer_opt = FeatureUnion([
    ("word_tfidf", word_vectorizer_opt),
    ("char_tfidf", char_vectorizer_opt),
])
```

```python
# Helper funct to fit a pipeline and print macro-F1 on the validation set

def evaluate_model(name, clf, X_train, y_train, X_valid, y_valid, use_combined_vec=False):
    '''
    name: string name to print
    clf: sklearn classifier
    use_combined_vec: if True, use word+char TF-IDF; otherwise only word-level
    '''
    if use_combined_vec:
        vec = combined_vectorizer
    else:
        vec = word_vectorizer

    pipe = Pipeline([
        ("vec", vec),
        ("clf", clf),
    ])

    pipe.fit(X_train, y_train)
    preds = pipe.predict(X_valid)
    f1 = f1_score(y_valid, preds, average="macro")

    print(f"=== {name} ===")
    print(f"Macro F1 on validation: {f1:.4f}")
    print()

    return pipe, f1, preds
```

## 3. Model Implementation Steps

```python
'''
Model 1 : word TF-IDF + Logistic Regression

'''

logreg_clf = LogisticRegression(
    max_iter=1000,
    n_jobs=-1,
)

logreg_pipe, f1_logreg, preds_logreg = evaluate_model(
    name="LogReg + word TF-IDF",
    clf=logreg_clf,
    X_train=X_train_text,
    y_train=y_train,
    X_valid=X_valid_text,
    y_valid=y_valid,
    use_combined_vec=False,  #(only words here)
)

print("Classification report (LogReg baseline):")
print(classification_report(y_valid, preds_logreg))
```

```python
# Model 2: word + char TF-IDF + Linear SVM

svm_clf = LinearSVC()

svm_pipe, f1_svm, preds_svm = evaluate_model(
    name="Linear SVM + word+char TF-IDF",
    clf=svm_clf,
    X_train=X_train_text,
    y_train=y_train,
    X_valid=X_valid_text,
    y_valid=y_valid,
    use_combined_vec=True,  # use combined word + char features
)

print("Classification report (Linear SVM):")
print(classification_report(y_valid, preds_svm))
```

```python
# Model 3: word TF-IDF + Complement Naive Bayes, cause it worked pretty well on lab 1

nb_clf = ComplementNB()

nb_pipe, f1_nb, preds_nb = evaluate_model(
    name="Complement NB + word TF-IDF",
    clf=nb_clf,
    X_train=X_train_text,
    y_train=y_train,
    X_valid=X_valid_text,
    y_valid=y_valid,
    use_combined_vec=False,
)

print("Classification report (Complement NB):")
print(classification_report(y_valid, preds_nb))
```

```python
# Compare F1 scores and pick the best model!!

scores = {
    "logreg_word": f1_logreg,
    "svm_word_char": f1_svm,
    "nb_word": f1_nb,
}

print("Validation macro-F1 scores:")
for name, score in scores.items():
    print(f"{name:15s}: {score:.4f}")

best_name = max(scores, key=scores.get)
print(f"\nBest model on this split: {best_name}")

# choose the corresponding fitted pipeline
if best_name == "logreg_word":
    best_pipe = logreg_pipe
elif best_name == "svm_word_char":
    best_pipe = svm_pipe
else:
    best_pipe = nb_pipe
```

```python
# Refit the chosen model on ALL labeled training data now

X_full=train_df["text_clean"]
y_full=train_df["emotion"]

# we need the same classifier type and vectorizer configuration as best_pipe
# but re-fitted on the full dataset

if best_name == "logreg_word":
    vec = word_vectorizer
    clf = LogisticRegression(max_iter=1000, n_jobs=-1)
    use_combined = False
elif best_name == "svm_word_char":
    vec = combined_vectorizer
    clf = LinearSVC()
    use_combined = True
else:
    vec = word_vectorizer
    clf = ComplementNB()
    use_combined = False

final_pipe = Pipeline([
    ("vec", vec),
    ("clf", clf),
])

final_pipe.fit(X_full, y_full)

print("Final model trained on full training data.")
```

```python
# Predict emotions for the test set and build submission.csv

test_preds = final_pipe.predict(test_df["text_clean"])

submission = pd.DataFrame({
    "id": test_df["id"],
    "emotion": test_preds,
})

print(submission.head())
print("Unique predicted labels:", submission["emotion"].unique())

submission.to_csv("submission.csv", index=False)
print("\nSaved submission.csv")
```
